```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf
from tensorflow import keras
```

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
from sklearn.datasets import make_moons
```

```python
x,y = make_moons(n_samples = 250,noise = 0.2,random_state = 8)
```

```python
x
```

```python
y
```

```python
plt.scatter(x[:,0],x[:,1],c=y,s=100)
```

```python
model = Sequential()
model.add(Dense(10,activation = "sigmoid",input_dim = 2))
model.add(Dense(10,activation = "sigmoid"))
model.add(Dense(10,activation = "sigmoid"))
model.add(Dense(10,activation = "sigmoid"))
model.add(Dense(10,activation = "sigmoid"))
model.add(Dense(10,activation = "sigmoid"))
model.add(Dense(10,activation = "sigmoid"))
model.add(Dense(10,activation = "sigmoid"))
model.add(Dense(10,activation = "sigmoid"))
model.add(Dense(10,activation = "sigmoid"))
model.add(Dense(10,activation = "sigmoid"))
model.add(Dense(1,activation = "sigmoid"))
```

```python
model.compile(loss = "binary_crossentropy",optimizer = "adam",metrics = ["accuracy"])
```

```python
model.get_weights()[0]
```

```python
old_weights = model.get_weights()[0]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2,random_state=4)
```

```python
model.fit(x_train,y_train,validation_data=(x_test,y_test),epochs=100)
```

If you see even after 100 epochs loss value hasnt decreased . this is vanishing gradient problem and hence we need to solve it

```python
new_weights = model.get_weights()[0]
```

```python
#approach 1 - reduce complexity
```

```python
model1 = Sequential()
model1.add(Dense(10,activation = "sigmoid",input_dim = 2))
model1.add(Dense(10,activation = "sigmoid"))
model1.add(Dense(1,activation = "sigmoid"))

model1.compile(loss = "binary_crossentropy",optimizer = "adam",metrics = ["accuracy"])

model1.fit(x_train,y_train,validation_data=(x_test,y_test),epochs=100)
```

if we reduce complexityy we see loss value has reduce significantly and accuracy is also increased to 80 and 88 %

```python
#approach 2 - using relu
```

```python
model3 = Sequential()
model3.add(Dense(10,activation = "relu",input_dim = 2))
model3.add(Dense(10,activation = "relu"))
model3.add(Dense(10,activation = "relu"))
model3.add(Dense(10,activation = "relu"))
model3.add(Dense(10,activation = "relu"))
model3.add(Dense(10,activation = "relu"))
model3.add(Dense(10,activation = "relu"))
model3.add(Dense(10,activation = "relu"))
model3.add(Dense(10,activation = "relu"))
model3.add(Dense(10,activation = "relu"))
model3.add(Dense(10,activation = "relu"))
model3.add(Dense(1,activation = "sigmoid"))

model3.compile(loss = "binary_crossentropy",optimizer = "adam",metrics = ["accuracy"])

model3.fit(x_train,y_train,validation_data=(x_test,y_test),epochs=100)
```

Using relu helps as it increase accuracy and reduces loss significantly
